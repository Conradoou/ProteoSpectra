```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# File path
file_path = r'E:\Correct_Data_Official\Official_Images\Coefficient.xlsx'

# Load the Excel file (specifying the sheet)
df = pd.read_excel(file_path)

# Display the first few rows to confirm the data structure
print(df.head())

# Select the columns for coefficients
columns = ['CTRL Coef.', 'TRT Coef.']

# Create DataFrame for the boxplot
boxplot_data = df.melt(value_vars=columns, var_name='Group', value_name='Correlation Coefficient')

# Define custom colors
palette = {'CTRL Coef.': '#FAAF62', 'TRT Coef.': '#99b6e0'}

# Create the boxplot
plt.figure(figsize=(4, 5.5))
sns.boxplot(x='Group', y='Correlation Coefficient', data=boxplot_data, palette=palette, width=0.4)

# Adjust labels and title
plt.title('Distribution of Correlation Coefficients')
plt.ylabel('Correlation Coefficient')
plt.xlabel('')

# Adjust layout
plt.tight_layout()

# Path to save the plots
save_path = r'E:\Correct_Data_Official\Official_Images\boxplot_coefficient'

# Save as TIFF and PNG
plt.savefig(f"{save_path}.tiff", dpi=300)
plt.savefig(f"{save_path}.png", dpi=300)

# Show the plot
plt.show()

```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# File path
file_path = r'E:\Correct_Data_Official\Official_Images\Coefficient.xlsx'

# Load the Excel file (automatically reading the first sheet)
df = pd.read_excel(file_path)

# Display the first few rows to confirm the data structure
print(df.head())

# Select the coefficient columns
columns = ['CTRL', 'TRT']

# Check if the columns exist in the DataFrame
for col in columns:
    if col not in df.columns:
        raise ValueError(f"The column '{col}' was not found in the Excel file.")

# Create DataFrame for the boxplot
boxplot_data = df.melt(value_vars=columns, var_name='Group', value_name='Correlation Coefficient')

# Define custom colors
palette = {'CTRL': '#FAAF62', 'TRT': '#99b6e0'}

# Create the boxplot without displaying outliers
plt.figure(figsize=(4, 5.5))
sns.boxplot(x='Group', y='Correlation Coefficient', data=boxplot_data, 
            palette=palette, width=0.4, showfliers=False)

# Adjust labels and title with a note about hidden outliers
plt.title('')
plt.ylabel('Coefficient of variation')
plt.xlabel('')

# Adjust layout
plt.tight_layout()

# Path to save the plots
save_path = r'E:\Correct_Data_Official\Official_Images\boxplot_coefficient'

# Save as TIFF and PNG
plt.savefig(f"{save_path}.tiff", dpi=300)
plt.savefig(f"{save_path}.png", dpi=300)

# Show the plot
plt.show()

```
